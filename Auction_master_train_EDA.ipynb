### 새로운 변수 생성

면적이 가격에 많은 영향을 끼친다고 고려되어 price를 land/area로 나눈 면적($m^2$) 별 가격 변수를 형성하도록 한다.<br/>

경매 가격에 관한 자료이므로 땅 면적,건물 관련 변수들 중 auction과 관련된 면적을 뽑아 변수를 만들었다.

```python
#땅 면적 관련 파생 변수
df['landappP']=df['Total_appraisal_price']/df['Total_land_auction_area']
df['landminP']=df['Minimum_sales_price']/df['Total_land_auction_area']
df['landhamP']=df['Hammer_price']/df['Total_land_auction_area']

#건물 면적 관련 파생 변수
df['buildingappP']=df['Total_appraisal_price']/df['Total_building_auction_area']
df['buildingminP']=df['Minimum_sales_price']/df['Total_building_auction_area']
df['buildinghamP']=df['Hammer_price']/df['Total_building_auction_area']
```

```python
#개인적으로 궁금한 파생변수를 만들어보았다
#1. 건물 전체 면적과  경매된 면적의 비율은 각자 어떻게될까?
df['building_auction_rate']=df['Total_building_auction_area']/df['Total_building_area']
#총 경매횟수에서 유찰된 비율은 얼마일까
df['miscarriage_rate']=df['Auction_miscarriage_count']/df['Auction_count']
```

```python
#이제 필요없는 변수 제거하자(변수 너무 많아...)
del df['Total_land_auction_area']; del df['Total_building_auction_area']; del df['Total_building_area']
del df['Total_land_gross_area']; del df['Total_land_real_area']
del df['Total_appraisal_price']; del df['Minimum_sales_price']
```

### 날짜 분석

```python
#날짜 뒤에 시간은 필요없으므로 제거한다.
df["First_auction_date_origin"]=df["First_auction_date"] #이에 관한 설명은 이후에 나온다.(날짜 길이 계산때)
df["Final_auction_date_origin"]=df["Final_auction_date"]

df["First_auction_date"]= df["First_auction_date"].str.split(" ", n = 1, expand = True)
df["Final_auction_date"]= df["Final_auction_date"].str.split(" ", n = 1, expand = True)

#https://www.geeksforgeeks.org/python-pandas-split-strings-into-two-list-columns-using-str-split/
```

```python
#요일 출력을 위해 ' '로 분리하고 datetime.strftime에 차례로 입력하여 요일을 출력해보도록한다.
from datetime import datetime
import numpy as np


#-를 기준으로 split
a_s=df["First_auction_date"].str.split("-",expand = True)    
a_e=df["Final_auction_date"].str.split("-", expand = True)


#split 결과가 object 형태로 반환되어 numeric으로 바꾸기
a_s[0]=pd.to_numeric(a_s[0]) 
a_s[1]=pd.to_numeric(a_s[1])
a_s[2]=pd.to_numeric(a_s[2])

a_e[0]=pd.to_numeric(a_e[0])
a_e[1]=pd.to_numeric(a_e[1])
a_e[2]=pd.to_numeric(a_e[2])


#요일 출력 변수 생성 및 초기값 지정
df['auction_start_weekday']=0
df['auction_end_weekday']=0


#아까 제외한 행들이 있으므로 index 다시 지정
df=df.reset_index()
#기존 index 행 제거(안해도됨, 필요없는 변수가 되고 index에 영향 없음)
del df['index']

#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html
```

```python
#요일 기록
for i in range(0,len(df)):
    df['auction_start_weekday'][i]=datetime(a_s[0][i], a_s[1][i], a_s[2][i]).strftime('%A')
    df['auction_end_weekday'][i]=datetime(a_e[0][i],a_e[1][i],a_e[2][i]).strftime('%A')

    
#https://stackoverflow.com/questions/9847213/how-do-i-get-the-day-of-week-given-a-date
```

```python
#다른 방식
import matplotlib.pyplot as plt
from matplotlib import gridspec

fig = plt.figure(figsize=(5, 10)) 
plt.tight_layout()
fig.subplots_adjust(bottom=0.5)

f,ax=plt.subplots(2,1) # 1x2 의 Figure 와 Axes

#각 subplot 크기를 조정하는 방법
gs = gridspec.GridSpec(nrows=2, # row 몇 개 
                       ncols=1, # col 몇 개 
                       height_ratios=[1, 1], 
                       width_ratios=[3]
                      )

ax[0] = plt.subplot(gs[0])
ax[0] = sns.kdeplot(df.query('Bid == 0')['landhamP'], color = "red",shade= True)
ax[0] = sns.kdeplot(df.query('Bid == 1')['landhamP'], color = 'blue', shade = True)
ax[0] = sns.kdeplot(df.query('Bid == 2')['landhamP'], color = 'green', shade = True)
ax[0].set_xlabel('낙찰가/토지 면적')
ax[0].set_ylabel('비율')
ax[0] = ax[0].legend(['일반','개별', '일괄'])

# 2번 plot
ax[1] = plt.subplot(gs[1])
ax[1] = sns.kdeplot(df.query('Bid == 0')['buildinghamP'], color = "red",shade= True)
ax[1] = sns.kdeplot(df.query('Bid == 1')['buildinghamP'], color = 'blue', shade = True)
ax[1] = sns.kdeplot(df.query('Bid == 2')['buildinghamP'], color = 'green', shade = True)
ax[1].set_xlabel('낙찰가/건물 면적')
ax[1].set_ylabel('비율')
ax[1] = ax[1].legend(['일반','개별', '일괄'])

plt.tight_layout()
plt.show()
```

낙찰가를 총 토지 경매 면적으로 나눈것이 건물 경매 면적으로 나눈 것보다 더 0에 가깝게 나옴을 알 수 있다.<br/>
특히나 Bid Class가 개별인 경우는 토기 경매 면적으로 나눈 것은 0에 가깝게 나오나 건물 면적으로 나눈 것은 1에 더욱더 가까이 나온다는 것을 알 수 있으며 묘한 쌍봉의 경태로 나옴을 알 수 있다.<br/>
